```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
df_iris = pd.read_csv('iris.csv')
```

```python
# change species names

df_iris = df_iris.replace('Iris-setosa','Setosa')
df_iris = df_iris.replace('Iris-versicolor','Versicolor')
df_iris = df_iris.replace('Iris-virginica','Virginica')
```

```python
df_iris.head()
```

```python
df_iris.tail()
```

```python
# choose/create color palette

flatui = ['#9b59b6', '#3498db', '#95a5a6', '#e74c3c', '#34495e', '#2ecc71']
sns.palplot(sns.color_palette(flatui))
```

```python
# distplot

sns.distplot(df_iris.SepalLengthCm);
```

The distplot()

```python
# swarmplot positions each scatterplot point on the categorical axis with an algorithm that avoids overlapping points

sns.swarmplot(x='Species', y='SepalLengthCm', data=df_iris);
```

```python
# swarmplot positions each scatterplot point on the categorical axis with an algorithm that avoids overlapping points

sns.swarmplot(x='SepalWidthCm', y='SepalLengthCm', hue='Species', data=df_iris);
```

The swarm plot shows the distribution and density of sepal length by species, with Iris virginica having the longest sepal length and setosa having the shortest.

```python
# a violinplot combines a boxplot with kernel density estimation 

sns.violinplot(x='Species', y='SepalLengthCm', data=df_iris);
```

The violin plot shows the density the distribution as well, in addition to showing the quartile ranges inside the violin.  The largest distribution for setosa sepal length is approximately 5 cm, versicolor is approximately 6cm and virginica is approximately 6.5cm.

```python
# combine swarm plot & violin plot to show each observation along with a summary of the distribution

sns.violinplot(x='Species', y='SepalLengthCm', data=df_iris, inner=None)
sns.swarmplot(x='Species', y='SepalLengthCm', data=df_iris, color = 'w', alpha=.5);
```

The combination of the swarm and box plots shows the same distributions overlapped.

```python
# bar plot

sns.barplot(x='Species', y='SepalLengthCm', data=df_iris);
```

The boxplot explicitly shows how the sepal length of each species compares to each other.  Virginica is the longest, and setosa is the shortest.

```python
df_iris.columns
```

```python
# boxplot with horizontal orientation

sns.boxplot(data=df_iris.iloc[:,1:], orient='h');
```

This boxplot shows the distribution by percentile for petallength/width and sepal length/width across all species.

```python
# bar plot

sns.barplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df_iris);
```

```python
sns.barplot(x='SepalWidthCm', y='SepalLengthCm', hue='Species', data=df_iris);
```

Grouped bar plots showing the difference in sepal length vs sepal width for each species.  The data looks completely different with sepal length being on the x-axis as shown in the first graph, versus sepal length being plotted on the y-axis in the second graph.

```python
sns.barplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df_iris);
```

```python
sns.barplot(x='PetalWidthCm', y='PetalLengthCm', hue='Species', data=df_iris);
```

This grouped bar plot for sepal length/width is much easier to read due to the consistency of the data distribution.  The data maintains the same pattern whether petal length is on the x-axis or the y-axis.  Each species is clearly identified.

```python
sns.set(style='ticks')

sns.pairplot(df_iris.iloc[:,1:], hue='Species');
```

Distributions of petal length/width and sepal length/width by species in scatter plots and stacked bar charts.  The joint plot shows multiple visualizations at a glance, which is insightful when trying to get an idea of how data relates to each other.

```python
# separate by species?? **multivariate color plot**

sns.jointplot(x='PetalLengthCm', y='PetalWidthCm', color = '#9b59b6', data=df_iris);
```

Combination scatter plot and histogram for petal length and petal width for all species with the pearsonr correlation coefficient.
