```python
# Making dummies;
# Create a data frame to hold features

features = pd.get_dummies(crime['murder'])

# create feature for murder to make categorical

features['murder_1'] = np.where(crime['murder']> 0, 1, 0)

print(pd.crosstab(features['murder_1'], crime['murder']))
```

```python
# create feature for robbery to make categorical

features['robbery_1'] = np.where(crime['robbery']>0, 1, 0)
print(pd.crosstab(features['robbery_1'], crime['robbery']))
```

```python
# create feature for pop_sq

features['pop_sq'] = crime.population **2
print(features['pop_sq'])
```

```python
corrmat = crime.corr()

# Set up the matplotlib figure

f, ax = plt.subplots(figsize=(12, 9))

# Draw the heatmap using seaborn

sns.heatmap(corrmat, vmax=.8, square=True)
plt.show()
```

```python
# In case some of the variables are on different scales than others (one ranges from 0 to 100, one ranges from 0 to
# 7, for example), scale them before averaging by subtracting the average of each variable from all values in 
# that variable, then dividing by the standard deviation.

means = crime[['population','violent_crime','murder','robbery','aggravated_assault','property_crime','burglary','larceny']].mean(axis=0)
stds = crime[['population','violent_crime','murder','robbery','aggravated_assault','property_crime','burglary','larceny']].std(axis=0)
features['theft'] = ((crime[['population','violent_crime','murder','robbery','aggravated_assault','property_crime','burglary','larceny']] - means) / stds).mean(axis=1)

# Check how well the composite correlates with each of the individual variables.

plotcrime = crime.loc[:, ['population','violent_crime','murder','robbery','aggravated_assault','property_crime','burglary','larceny']]
plotcrime['theft'] = features['theft'] 
corrmat2 = plotcrime.corr()

print(corrmat2)
```

**Based on the correlation matrix, there is high correlation between population, property crime, burglary, larceny.  This is expected since these types of crimes are similar in nature.  The higher the population, the more crimes committed.**
